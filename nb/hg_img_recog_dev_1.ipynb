```python
import requests
import os
from dotenv import load_dotenv

# setting token in colab notebook keys
from google.colab import userdata
HF_API_TOKEN= userdata.get('HF_API_KEY')


# # Load Hugging Face API Token from .env
# load_dotenv()
# HF_API_TOKEN = os.getenv("HF_API_KEY")


# Hugging Face ViT model
HF_MODEL_URL = "https://api-inference.huggingface.co/models/google/vit-base-patch16-224"

# Headers for Hugging Face API
HEADERS = {
    "Authorization": f"Bearer {HF_API_TOKEN}",
    "Content-Type": "application/octet-stream"
}

def download_image_from_url(image_url):
    print(f"üì• Downloading image from: {image_url}")
    response = requests.get(image_url)
    if response.status_code != 200:
        print("‚ùå Failed to download image")
        return None
    return response.content

def recognize_image_from_url(image_url):
    image_data = download_image_from_url(image_url)
    if image_data is None:
        return

    print("üîç Sending image to Hugging Face Inference API...")
    response = requests.post(HF_MODEL_URL, headers=HEADERS, data=image_data)

    if response.status_code != 200:
        print(f"‚ùå API call failed: {response.status_code}")
        print(response.text)
        return

    predictions = response.json()
    print("‚úÖ Recognition Results:")
    for pred in predictions:
        label = pred.get("label", "unknown")
        score = pred.get("score", 0)
        print(f"  ‚Ä¢ {label}: {score:.2%}")

if __name__ == "__main__":
    # Example image URL (you can replace this)
    image_url = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRS0IQhVr9DDJCq61QX28zCoiqDrvezBh5ylw&s"
    recognize_image_from_url(image_url)
```
